```python
# bar() IS USED TO CREATE VERTICAL BARS IN THE PLOT .
# PLOT VERTICAL BARS IN THE FIGURE .

import matplotlib.pyplot as plt
import numpy as np

student = np.array ([ "LEO" , "DAX" , "ENO" , "BOB" , "TOM" , "LUX" , "ZEN" , "ASH" ])
marks = np.array ([24,36,16,40,28,12,20,32])

plt.bar ( student , marks )
plt.show()
```

```python
# barh() IS USED TO CREATE HORIZONTAL BARS IN THE PLOT .
# PLOT HORIZONTAL BARS IN THE FIGURE .

import matplotlib.pyplot as plt
import numpy as np

student = np.array ([ "LEO" , "DAX" , "ENO" , "BOB" , "TOM" , "LUX" , "ZEN" , "ASH" ])
marks = np.array ([24,36,16,40,28,12,20,32])

plt.barh ( student , marks )
plt.show()
```

```python
# color ATTRIBUTE IS USED TO GIVE COLOR TO THE BAR .
# PLOT HORIZONTAL BARS WITH OWN COLOUR IN THE FIGURE .

import matplotlib.pyplot as plt
import numpy as np

student = np.array ([ "LEO" , "DAX" , "ENO" , "BOB" , "TOM" , "LUX" , "ZEN" , "ASH" ])
marks = np.array ([24,36,16,40,28,12,20,32])

plt.bar ( student , marks , color ="gold" )
plt.show()
```

```python
# PLOT HORIZONTAL BARS WITH DIFFERENT COLOUR IN EACH BAR IN THE FIGURE .

import matplotlib.pyplot as plt
import numpy as np

student = np.array ([ "LEO" , "DAX" , "ENO" , "BOB" , "TOM" , "LUX" , "ZEN" , "ASH" ])
marks = np.array ([24,36,16,40,28,12,20,32])

colours_list = [ "brown" , "aqua" , "purple" , "lime" , "peru" , "teal" , "orange" , "orchid" ]

plt.bar ( student , marks , color=colours_list )
plt.show()
```

```python
# width ATTRIBUTE IS USED TO CHANGE THE VERTICAL BAR WIDTH .
# THE DEFAULT VALUE  OF width IS 0.80 .
# PLOT VERTICAL BAR WITH BARS OF OWN WIDTH .

import matplotlib.pyplot as plt
import numpy as np

student = np.array ([ "LEO" , "DAX" , "ENO" , "BOB" , "TOM" , "LUX" , "ZEN" , "ASH" ])
marks = np.array ([24,36,16,40,28,12,20,32])

plt.bar ( student , marks , width=0.64 )
plt.show()
```

```python
# height ATTRIBUTE IS USED TO CHANGE THE HORIZONTAL BAR WIDTH .
# THE DEFAULT VALUE  OF height IS 0.80 .
# PLOT HORIZONTAL BAR WITH BARS OF OWN WIDTH .

import matplotlib.pyplot as plt
import numpy as np

student = np.array ([ "LEO" , "DAX" , "ENO" , "BOB" , "TOM" , "LUX" , "ZEN" , "ASH" ])
marks = np.array ([24,36,16,40,28,12,20,32])

plt.barh ( student , marks , height=0.56 )
plt.show()
```
