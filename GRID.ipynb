```python
# PLOT THE GRAPTH WITHOUT GRID LINES .

import matplotlib.pyplot as plt
import numpy as np

xx = np.array ([80,85,90,95,100,105,110,115,120,125])
yy = np.array ([240,250,260,270,280,290,300,310,320,330])

font_one = { "family":"serif" , "color":"orange" , "size":20 }
font_two = { "family":"serif" , "color":"brown" , "size":16 }

plt.plot ( xx , yy , marker="h" , markersize=10 )

plt.xlabel ( "AVERAGE PULSE" , fontdict = font_two )
plt.ylabel ( "CALORIE BURNAGE" , fontdict = font_two )
plt.title ( "EXERCISE DATA" , fontdict = font_one , loc="center" )

plt.show()
```
