```python
# hist() IS USED TO PLOT THE HISTOGRAM GRAPH .
# hist() IS A GRAPH SHOWING THE FREQUENCY DISTRIBUTION .

import numpy as np
ww = np.random.normal ( loc=160 , scale=4 , size=(20) )
print(ww)
```

```python
import numpy as np
xx = np.random.normal ( loc=100 , scale=10 , size=30 )
print(xx)
```

```python
import numpy as np
yy = np.random.normal ( 100 , 20 , 60 )
print(yy)
```

```python
import numpy as np
zz = np.random.normal ( 100 , 20 , (6,8) )
print(zz)
```

```python
# PLOT THE HISTOGRAM OF THE NORMAL DISTRIBUTION WITH bins ATTRIBUTE .

import matplotlib.pyplot as plt
import numpy as np

xyz = np.random.normal ( loc=100 , scale=16 , size=64 )
plt.hist ( xyz , bins=8 )
plt.show()
```

```python
# PLOT THE HISTOGRAM OF THE NORMAL DISTRIBUTION WITHOUT bins ATTRIBUTE .

import matplotlib.pyplot as plt
import numpy as np

xyz = np.random.normal ( loc=100 , scale=16 , size=64 )
plt.hist ( xyz )
plt.show()
```

```python
# PLOT THE HISTOGRAM OF THE NORMAL DISTRIBUTION WITH width ATTRIBUTE .

import matplotlib.pyplot as plt
import numpy as np

xyz = np.random.normal ( loc=100 , scale=16 , size=64 )
plt.hist ( xyz , width=6 )
plt.show()
```

```python
# PLOT THE HISTOGRAM OF THE NORMAL DISTRIBUTION WITH color ATTRIBUTE .

import matplotlib.pyplot as plt
import numpy as np

xyz = np.random.normal ( loc=100 , scale=16 , size=64 )
plt.hist ( xyz , color="peru" )
plt.show()
```
