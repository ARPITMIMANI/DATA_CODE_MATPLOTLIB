```python
# xlabel() AND ylabel() ARE USED TO GIVE LABEL TO THE X-AXIS AND Y-AXIS RESPECTIVELY IN THE PLOT .
# title() IS USED TO GIVE TITLE TO THE GRAPH PLOT .
# PLOT THE GRAPH WITH LABEL AND TITLE .

import matplotlib.pyplot as plt
import numpy as np

x = np.array ([80,85,90,95,100,105,110,115,120,125])
y = np.array ([240,250,260,270,280,290,300,310,320,330])

plt.plot ( x , y , marker="h" , markersize=10 )

plt.xlabel("AVERAGE PULSE")
plt.ylabel("CALORIE BURNAGE")
plt.title("EXERCISE DATA")

plt.grid()
plt.show()
```

```python
# fontdict ARGUMENT IS USED TO SET THE FONT PROPERTY FOR TITLE AND LABELS .

import matplotlib.pyplot as plt
import numpy as np

x = np.array ([80,85,90,95,100,105,110,115,120,125])
y = np.array ([240,250,260,270,280,290,300,310,320,330])

font_one = { "family":"serif" , "color":"orange" , "size":20 }
font_two = { "family":"serif" , "color":"brown" , "size":16 }

plt.plot ( x , y , marker="h" , markersize=10 )

plt.xlabel ( "AVERAGE PULSE" , fontdict = font_two )
plt.ylabel ( "CALORIE BURNAGE" , fontdict = font_two )
plt.title ( "EXERCISE DATA" , fontdict = font_one )

plt.grid()
plt.show()
```

```python
# loc ARGUMENT IS USED TO CHANGE THE LOCATION OF TITLE IN THE PLOT .

import matplotlib.pyplot as plt
import numpy as np

x = np.array ([80,85,90,95,100,105,110,115,120,125])
y = np.array ([240,250,260,270,280,290,300,310,320,330])

font_one = { "family":"serif" , "color":"orange" , "size":20 }
font_two = { "family":"serif" , "color":"brown" , "size":16 }

plt.plot ( x , y , marker="h" , markersize=10 )

plt.xlabel ( "AVERAGE PULSE" , fontdict = font_two )
plt.ylabel ( "CALORIE BURNAGE" , fontdict = font_two )
plt.title ( "EXERCISE DATA" , fontdict = font_one , loc="center" )

plt.grid()
plt.show()
```
