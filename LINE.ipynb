```python
# linestyle OR ls ARGUMENT IS USED TO CHANGE THE STYLE OF THE PLOTTED LINE .

import matplotlib.pyplot as plt
import numpy as np

xyz = np.array ([24,36,12,40,16,4,28,20,8,32])
plt.plot ( xyz , linestyle = ":" )

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xyz = np.array ([24,36,12,40,16,4,28,20,8,32])
plt.plot ( xyz , ls = "dashdot" )

plt.grid()
plt.show()
```

```python
# color OR c ARGUMENT IS USED TO CHANGE THE STYLE OF THE PLOTTED LINE .

import matplotlib.pyplot as plt
import numpy as np

xyz = np.array ([24,36,12,40,16,4,28,20,8,32])
plt.plot ( xyz , color = "orange" )

plt.grid()
plt.show()
```

```python
# linewidth OR lw ARGUMENT IS USED TO CHANGE THE WIDTH OF THE PLOTTED LINE .

import matplotlib.pyplot as plt
import numpy as np

xyz = np.array ([24,36,12,40,16,4,28,20,8,32])
plt.plot ( xyz , linewidth=4 )

plt.grid()
plt.show()
```
