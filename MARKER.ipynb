```python
# marker ARGUMENT IS USED TO EMPHASIZE EACH POINT WITH A SPECIFIC MARKER .

import matplotlib.pyplot as plt
import numpy as np

xyz = np.array ([24,36,12,40,16,4,28,20,8,32])
plt.plot ( xyz , marker="d" )

plt.grid()
plt.show()
```

```python
''' 

MARKER LIST ARE : "1" , "2" , "3" , "4" , "8" , "p" , "P" , "h" , "H" , "x" , "X" , "d" , "D" , "s" , "o" , 
"v" , "^" , ">" , "<" , "-" , "|" , "*" , "+" , "." , "," . 

LINE STYLE LIST ARE : "-" SOLID LINE , ":" DOTTED LINE , "--" DASHED LINE , "-." DASHDOT LINE .

COLOR REFERENCES ARE : "b" BLUE , "c" CYAN , "g" GREEN , "k" BLACK , "m" MAGENTA , "r" RED , "w" WHITE , "y" YELLOW .

'''

# FORMAT STRINGS "fmt" ALSO KNOWN AS STRING NOTATION PARAMETER DENOTED BY "marker|line|color" .

import matplotlib.pyplot as plt
import numpy as np

xyz = np.array ([24,36,12,40,16,4,28,20,8,32])
plt.plot ( xyz , "*--g" )

plt.grid()
plt.show()
```

```python
# markersize OR ms ARGUMENT IS USED TO CHANGE THE MARKER SIZE .

import matplotlib.pyplot as plt
import numpy as np

xyz = np.array ([24,36,12,40,16,4,28,20,8,32])
plt.plot ( xyz , marker="h" , ms=10 )

plt.grid()
plt.show()
```

```python
# mec ARGUMENT IS USED TO CHANGE THE MARKER EDGE COLOR .

import matplotlib.pyplot as plt
import numpy as np

xyz = np.array ([24,36,12,40,16,4,28,20,8,32])
plt.plot ( xyz , marker="s" , ms=6 , mec= "m" )

plt.grid()
plt.show()
```

```python
# mfc ARGUMENT IS USED TO CHANGE THE MARKER FACE COLOR .

import matplotlib.pyplot as plt
import numpy as np

xyz = np.array ([24,36,12,40,16,4,28,20,8,32])
plt.plot ( xyz , marker="s" , ms=6 , mec="m" , mfc="y" )

plt.grid()
plt.show()
```
