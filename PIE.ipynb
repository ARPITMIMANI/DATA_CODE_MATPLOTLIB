```python
# pie() IS USED TO CREAT PIE CHARTS .

import matplotlib.pyplot as plt
import numpy as np

fruit_numbers = np.array ([16,32,8,24,20])
plt.pie(fruit_numbers)
plt.show()
```

```python
# labels ATTRIBUTE IS USED TO LABEL THE PIE CHART .

import matplotlib.pyplot as plt
import numpy as np

fruit_numbers = np.array ([16,32,8,24,20])
fruits_list = np.array ([ "MANGO" , "BANANA" , "ORANGE" , "APPLE" , "WATERMELON" ])

plt.pie ( fruit_numbers , labels=fruits_list )
plt.show()
```

```python
# startangle ARGUMENT IS USED FOR CUSTOM START ANGLE .
# THE DEFAULT startangle VALUE IS "0" WHICH IS THE X-AXIS .

import matplotlib.pyplot as plt
import numpy as np

fruit_numbers = np.array ([16,32,8,24,20])
fruits_list = np.array ([ "MANGO" , "BANANA" , "ORANGE" , "APPLE" , "WATERMELON" ])

plt.pie ( fruit_numbers , labels=fruits_list , startangle=+90 )
plt.show()
```

```python
# explode ATTRIBUTE IS USED FOR OFFSETTING A SLICE . 

import matplotlib.pyplot as plt
import numpy as np

fruit_numbers = np.array ([16,32,8,24,20])
fruits_list = np.array ([ "MANGO" , "BANANA" , "ORANGE" , "APPLE" , "WATERMELON" ])
fruit_explode_one = [ 0.2 , 0 , 0 , 0 , 0 ]

plt.pie ( fruit_numbers , labels = fruits_list , explode = fruit_explode_one )
plt.show()
```

```python
# shadow ATTRIBUTE IS USED FOR ADDING A DROP SHADOW TO THE PIE CHART .

import matplotlib.pyplot as plt
import numpy as np

fruit_numbers = np.array ([16,32,8,24,20])
fruits_list = np.array ([ "MANGO" , "BANANA" , "ORANGE" , "APPLE" , "WATERMELON" ])

plt.pie ( fruit_numbers , labels=fruits_list , shadow=True )
plt.show()
```

```python
# colors IS USED TO SPECIFY CUSTOM COLOUR TO EACH SLICE .

import matplotlib.pyplot as plt
import numpy as np

fruit_numbers = np.array ([16,32,8,24,20])
fruits_list = np.array ([ "MANGO" , "BANANA" , "ORANGE" , "APPLE" , "WATERMELON" ])
fruit_colours = np.array ([ "sandybrown" , "olive" , "navy" , "darkviolet" , "brown" ])

plt.pie ( fruit_numbers , labels=fruits_list , colors=fruit_colours )
plt.show()
```

```python
# legend() IS USED TO GIVE LIST OF EXPLAINATION ABOUT THE PIE CHART .

import matplotlib.pyplot as plt
import numpy as np

fruit_numbers = np.array ([16,32,8,24,20])
fruits_list = np.array ([ "MANGO" , "BANANA" , "ORANGE" , "APPLE" , "WATERMELON" ])
fruit_colours = np.array ([ "sandybrown" , "olive" , "navy" , "darkviolet" , "brown" ])

plt.pie ( fruit_numbers , labels=fruits_list , colors=fruit_colours )
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fruit_numbers = np.array ([16,32,8,24,20])
fruits_list = np.array ([ "MANGO" , "BANANA" , "ORANGE" , "APPLE" , "WATERMELON" ])
fruit_colours = np.array ([ "sandybrown" , "olive" , "navy" , "darkviolet" , "brown" ])

plt.pie ( fruit_numbers , labels=fruits_list , colors=fruit_colours )
plt.legend ( loc="upper right" )
plt.show()
```

```python
# title IS USED TO ADD LEGEND WITH HEADER .

import matplotlib.pyplot as plt
import numpy as np

fruit_numbers = np.array ([16,32,8,24,20])
fruits_list = np.array ([ "MANGO" , "BANANA" , "ORANGE" , "APPLE" , "WATERMELON" ])
fruit_colours = np.array ([ "sandybrown" , "olive" , "navy" , "darkviolet" , "brown" ])

plt.pie ( fruit_numbers , labels=fruits_list , colors=fruit_colours )
plt.legend ( title="FRUITS" )
plt.show()
```

```python
# radius ATTRIBUTE IS USED TO CHANGE THE RADIUS OF THE PIE CHART .

import matplotlib.pyplot as plt
import numpy as np

fruit_numbers = np.array ([16,32,8,24,20])
fruits_list = np.array ([ "MANGO" , "BANANA" , "ORANGE" , "APPLE" , "WATERMELON" ])
fruit_colours = np.array ([ "sandybrown" , "olive" , "navy" , "darkviolet" , "brown" ])

plt.pie ( fruit_numbers , labels=fruits_list , colors=fruit_colours , radius=1.80 )
plt.legend ( title="FRUITS" )
plt.show()
```
