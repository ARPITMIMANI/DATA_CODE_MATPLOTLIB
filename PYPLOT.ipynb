```python
# plot() IS USED TO DRAW THE POINTS OR MARKERS IN A DIAGRAM .
# THERE ARE 2 PARAMETERS TO SPECIFY POINTS IN THE DIAGRAM : X-AXIS AND Y-AXIS .

# REFER pyplot PACKAGE AS plt .
# PLOT THE GRAPH WITH LINE .

import matplotlib.pyplot as plt
import numpy as np

a = np.array ([1,8])
b = np.array ([3,10])

plt.plot(a,b)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

c = np.array ([11,14,17,20,23,26,29,32])
d = np.array ([10,20,30,40,50,60,70,80])

plt.plot(c,d)
plt.grid()
plt.show()
```

```python
# THE X-AXIS IS HORIZONTAL AXIS .
# THE Y-AXIS IS VERTICAL AXIS .

# PLOT WITHOUT LINE .

import matplotlib.pyplot as plt
import numpy as np

a = np.array ([1,8])
b = np.array ([3,10])

plt.plot ( a , b , "o" )
plt.grid()
plt.show()
```

```python
# PLOT THE ZIG ZAG PLOT .

import matplotlib.pyplot as plt
import numpy as np

e = np.array ([1,2,3,4,5,6,7,8])
f = np.array ([4,6,3,2,8,5,1,7])

plt.plot ( e , f , "-" )
plt.grid()
plt.show()
```

```python
# PLOTTING WITH ONLY Y-AXIS POINTS .
# IF WE SPECIFY THE POINTS OF X-AXIS THEN THE DEFAULT VALUES OF X-AXIS WILL BE 0 , 1 , 2 , 3 , 4 ..... WILL APPLY .

import matplotlib.pyplot as plt
import numpy as np

g = np.array ([14,16,11,15,13,12])

plt.plot(g)
plt.grid()
plt.show()
```
