```python
# subplot() IS USED TO DRAW MULTIPLE PLOTS IN ONE FIGURE .

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array ([31,32,33,34,35,36,37,38])
y1 = np.array ([14,17,12,18,13,11,16,15])

plt.subplot(1,2,1)
plt.plot(x1,y1)

# THIS FIGURE HAS 1 ROW 2 COLUMNS AND IS THE 1ST PLOT .

x2 = np.array ([1,2,3,4,5,6,7,8])
y2 = np.array ([25,23,28,21,24,27,22,26])

plt.subplot(1,2,2)
plt.plot(x2,y2)

# THIS FIGURE HAS 1 ROW 2 COLUMNS AND IS THE 2ND PLOT .

plt.show()
```

```python
# suptitle() IS USED TO GIVE SUPER TITLE TO THE FIGURE .
# PLOT 6 PLOTS IN ONE FIGURE WITH TITLE AND WITH SUPER TITLE .

import matplotlib.pyplot as plt
import numpy as np

x3 = np.array ([1,2,3,4,5,6,7,8])
y3 = np.array ([48,16,27,32,54,45,12,58])

plt.subplot(2,3,1)
plt.plot(x3,y3)
plt.title("MONTH_ONE")

# THIS FIGURE HAS 2 ROWS 3 COLUMNS AND IS THE 1ST PLOT .

x4 = np.array ([1,2,3,4,5,6,7,8])
y4 = np.array ([18,29,46,24,39,56,60,19])

plt.subplot(2,3,2)
plt.plot(x4,y4)
plt.title("MONTH_TWO")

# THIS FIGURE HAS 2 ROWS 3 COLUMNS AND IS THE 2ND PLOT .

x5 = np.array ([1,2,3,4,5,6,7,8])
y5 = np.array ([36,42,15,22,53,35,11,31])

plt.subplot(2,3,3)
plt.plot(x5,y5)
plt.title("MONTH_THREE")

# THIS FIGURE HAS 2 ROWS 3 COLUMNS AND IS THE 3RD PLOT .

x6 = np.array ([1,2,3,4,5,6,7,8])
y6 = np.array ([47,17,30,44,14,21,19,55])

plt.subplot(2,3,4)
plt.plot(x6,y6)
plt.title("MONTH_FOUR")

# THIS FIGURE HAS 2 ROWS 3 COLUMNS AND IS THE 4TH PLOT .

x7 = np.array ([1,2,3,4,5,6,7,8])
y7 = np.array ([43,37,20,38,57,15,28,23])

plt.subplot(2,3,5)
plt.plot(x7,y7)
plt.title("MONTH_FIVE")

# THIS FIGURE HAS 2 ROWS 3 COLUMNS AND IS THE 5TH PLOT .

x8 = np.array ([1,2,3,4,5,6,7,8])
y8 = np.array ([51,34,26,13,52,41,15,28])

plt.subplot(2,3,6)
plt.plot(x8,y8)
plt.title("MONTH_SIX")

# THIS FIGURE HAS 2 ROWS 3 COLUMNS AND IS THE 6TH PLOT .

plt.suptitle("SALES REPORT")
plt.show()
```
